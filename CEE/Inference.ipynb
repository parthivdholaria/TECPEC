```python
from simpletransformers.question_answering import QuestionAnsweringModel
import json
import torch
import pickle
```

### Inference

```python
val_args = {
        'fp16': False,
        'overwrite_output_dir': False, 
        'doc_stride': 512, 
        'max_query_length': 512, 
        'max_answer_length': 512,
        "max_seq_length": 200,
        'n_best_size': 1,
        'null_score_diff_threshold': 0.0,
        'sliding_window': False,
        'eval_batch_size': 16
    }
```

```python
val_data_pred = json.load(open('/kaggle/input/semeval3-task-3-dataset/Dataset/CEE_QA/Predicted_Emotions/RoBERTa/val_utterance_level_roberta_pred.json'))
```

```python
qa_model = QuestionAnsweringModel('bert', '/kaggle/input/causeqa/pytorch/causeqa/1/CauseQA', args=val_args, cuda_device=1)
```

```python
answer_list, probability_list = qa_model.predict(val_data_pred)
```

```python
json.dump(answer_list, open('/kaggle/working/val_cause_pred_roberta.json', 'w'), indent=4)
```
