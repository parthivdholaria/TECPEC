```python
from simpletransformers.question_answering import QuestionAnsweringModel
import json
import torch
```

```python
train_data = json.load(open('/kaggle/input/Dataset/CEE_QA/train_simple_transformers.json'))
val_data = json.load(open('/kaggle/input/Dataset/CEE_QA/val_simple_transformers.json'))
```

```python
!mkdir /kaggle/working/best_model
```

```python
train_args = {
    'fp16': False,
    'overwrite_output_dir': True, 
    'doc_stride': 512, 
    'max_query_length': 512, 
    'max_answer_length': 200,
    "max_seq_length": 512,
    'n_best_size': 20,
    'null_score_diff_threshold': 0.0,
    'learning_rate': 1e-4,
    'sliding_window': False,
    'output_dir': "/kaggle/working/",
    'best_model_dir': "/kaggle/working/" + 'best_model/',
    'evaluate_during_training': True,
    'evaluate_during_training_steps': int(48697/4),
    'save_eval_checkpoints': False,
    'save_model_every_epoch': False,
    'save_steps': 500000,
    'train_batch_size': 16,
    'num_train_epochs': 3,
}
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
device
```

```python
qa_model = QuestionAnsweringModel('bert', 'mrm8488/spanbert-finetuned-squadv2', args=train_args, cuda_device=1, use_cuda=True)
qa_model.train_model(train_data, eval_data=val_data)
```
